```python
from sklearn.datasets import make_blobs, load_digits
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import matplotlib.pyplot as plt
import pandas as pd
```

### Load data

```python
centers = [[1, 1], [-1, -1], [1, -1]]

X, labels = make_blobs(n_samples=750, centers=centers, cluster_std=0.4, random_state=42)

X
```

### Plot the blobs

```python
plt.scatter(X[:, 0], X[:, 1])
```

### Run k-means

```python
# define the model
kmeans = KMeans(n_clusters=3)

# run the model
kmeans.fit(X)
```

```python
# look at labels from kmeans
# kmeans.labels_
```

```python
# plot with colors from kmeans clustering
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_)
```

### Run DBSCAN

```python
# define the model
dbscan = DBSCAN(eps=0.3, min_samples=10)

# run the model
dbscan.fit(X)

# plot the results
plt.scatter(X[:, 0], X[:, 1], c=dbscan.labels_)
```

### Run agglomerative clustering

```python
# define the model
agg = AgglomerativeClustering(n_clusters=3)

# run the model
agg.fit(X)

# plot the result
plt.scatter(X[:, 0], X[:, 1], c=agg.labels_)
```

### Now apply clustering to the digits dataset

### Load data

```python
digits = load_digits()

X = digits.data
y = digits.target

X
```

### Do dimension reduction for plotting

```python
lda = LDA(n_components=2)

X_lda = lda.fit_transform(X, y)

X_lda 
```

### Plot the dimension reduced data

```python
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, alpha=0.6)
plt.colorbar(label="Digit")
```

### Run k-means

```python
# define the model
kmeans = KMeans(n_clusters=10, random_state=42)

# run the model
kmeans.fit(X)
```

```python
# evaluate the clustering
kmeans_score = adjusted_rand_score(y, kmeans.labels_)

kmeans_score
```

### Run DBSCAN on digits

### Run loop to test different values of n_clusters for K means

```python
num_cluster_list = [6, 8, 9, 10, 11, 12, 14]

for num in num_cluster_list:
    kmeans = KMeans(n_clusters=num, random_state=42)

    kmeans.fit(X)

    score = adjusted_rand_score(y, kmeans.labels_)

    print(num, score)
```

```python
eps_list = [5, 10, 15, 25, 40]

for num in eps_list:
    dbscan = DBSCAN(eps=num, min_samples=10)

    dbscan.fit(X)

    score = adjusted_rand_score(y, dbscan.labels_)

    print(num, score)
```

```python
num_cluster_list = [6, 8, 9, 10, 11, 12, 14]

for num in num_cluster_list:
    agg = AgglomerativeClustering(n_clusters=num)

    agg.fit(X)

    score = adjusted_rand_score(y, agg.labels_)

    print(num, score)
```
